```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image =plt.imread('pepo_evil.jpg')
plt.imshow(image)
```

```python
greyImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(greyImage, cmap='gray')
```

```python
redHist = cv2.calcHist([image[:,:,0]], [0], None, [256], [0, 256])
plt.plot(redHist, color='red')
greenHist = cv2.calcHist([image[:,:,1]], [0], None, [256], [0, 256])
plt.plot(greenHist, color='green')
blueHist = cv2.calcHist([image[:,:,2]], [0], None, [256], [0, 256])
plt.plot(blueHist, color='blue')
```
